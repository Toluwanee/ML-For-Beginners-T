```python
import pandas as pd

#Read the file
pumpkins = pd.read_csv('../data/US-pumpkins.csv')

#Filter out rows that don't use price per bushel
pumpkins = pumpkins[pumpkins['Package'].str.contains('bushel',case=True, regex=True)]
pumpkins.head()
```

```python
# Count how many cells in each column are empty
pumpkins.isnull().sum()
```

```python
# Define columns to keep
new_columns = ["Package", "Month", "Low Price", "High Price", "Date"]

# Drop all other columns
pumpkins = pumpkins.drop([c for c in pumpkins.columns if c not in new_columns], axis=1)
```

```python
# Calculate the average price from the high and low price columns
price = (pumpkins["Low Price"] + pumpkins["High Price"]) / 2 

# Get month from date column
month = pd.DatetimeIndex(pumpkins["Date"]).month

# Create new data frames with the new columns
new_pumpkins = pd.DataFrame({ "Month": month, "Package": pumpkins ["Package"], "Low Price": pumpkins["Low Price"], "High Price": pumpkins["High Price"]  })
```

```python
# Normalise price of bushel
# Convert price of all cells prices by 1 1/9 bushels by dividing by 1/9
new_pumpkins.loc[new_pumpkins["Package"].str.contains("1 1/9"), "Price"] = price/(1 + 1/9)

# Normalise() Convert the price of all cells prices by 1/2 bushels by dividing by 1/2
new_pumpkins.loc[new_pumpkins["Package"].str.contains("1/2"), "Price"] = price/(1/2)
```

```python
#print the first few rows of the new data frame
new_pumpkins.head()
```

```python
new_pumpkins.tail()
```

```python
import matplotlib.pyplot as plt

# Get the values to plot
price = new_pumpkins.Price
month = new_pumpkins.Month

#Crreate and show scatter plot of omth against price
plt.scatter(price, month)
plt.show()
```

```python
#Group the pumpkins into a bar chart
new_pumpkins.groupby(["Month"])["Price"].mean().plot(kind="bar")

#Add a label to the bar chart
plt.ylabel("Pumpkin Price")
```

The best time to buy pumpkins is december. The pumpkins are more expensive in September and October. Why could that been Halloween is celebrated on the last day in October and pumpkins are used for decorations and costume.
